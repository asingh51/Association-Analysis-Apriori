```python
# Import your Libraries
import apriori as ap
import csv
import pandas as pd
```

```python
# # # # # # # # #  Code for Task 1 # # # # # # # # #
data=pd.read_csv('dataset_group.csv', names = ["Date", "TrNo", "Item"])
dataset=data.drop_duplicates()
transactionDataset= dataset.groupby('TrNo')['Item'].apply(list)
transactionDataset.to_csv('dataset.csv',header=False)

with open('dataset.csv', "w") as file:
    writer = csv.writer(file)
    for row in transactionDataset:
        writer.writerow(row)
file.close()

print('First 7 transactions of the dataset.csv')
pd.set_option('display.max_colwidth',-1)
dataset2=pd.read_csv('dataset.csv', sep='delimiter', engine='python', names=['Items'])
dataset2.head(7)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:1 # # # # # # # # #') 
print ('Case 1 (minimum support=0.33 and minimum confidence=0.1)')
print ('Case 1 Reasoning: This represents the Best case: looing at the rule. with this, we can consider that people who buys poultry also buys Vegetables with .787 confidence. This is the most common rule with high reoccouring item in transaction and high chances of association purchase. ')
print ('Therefore, vegetables and poultry must be placed side by side to increase the profit.')
print ('Case 1 Output:')
items, rules = ap.runApriori(ap.dataFromFile('dataset.csv'), 0.31, 0.1)
ap.printResults(items, rules)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:2 # # # # # # # # #') 
print ('Case 2 (minimum support=0.1 and minimum confidence=0.9)') 
print ('Case 2 Reasoning: In the second case, I have tried to focus on association rather than more repetition of tranaction. By keeping the support low and high confidence means that the one who buys eggs and sandwich loaves most of the times buys vegetables')
print ('It shows a strong association chances of purchase.')
print ('Case 2 Output:') 
items, rules = ap.runApriori(ap.dataFromFile('dataset.csv'), 0.1, 0.9)
ap.printResults(items, rules)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:3 # # # # # # # # #') 
print ('Case 3 (minimum support=0.2 and minimum confidence=0.4)')
print ('Case 3 Reasoning: Third case represent most of the general cases of possible and available ')
print ('combinations which can be used in the store for the association. By keeping every value average, we can can see the general rules')
print ('Case 3 Output:')
items, rules = ap.runApriori(ap.dataFromFile('dataset.csv'), 0.2, 0.4)
ap.printResults(items, rules)
```
